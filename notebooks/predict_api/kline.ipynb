```python
from datetime import datetime, timedelta

now = datetime.now()
three_days_ago = timedelta(days=2)

three_days_ago_timestamp = int(datetime.timestamp(now - three_days_ago))
three_days_ago_timestamp
```

```python
from pybit import HTTP
from datetime import datetime, timedelta
from time import time

session = HTTP("https://api.bybit.com")
res = session.query_kline(symbol="BTCUSDT", interval="15", from_time=f"{three_days_ago_timestamp}")
res
```

```python
import pandas as pd
df = pd.DataFrame(res["result"])
df.head()
```

```python
df = df.set_index("open_time")
df = df.sort_index()
df.head()
```

```python
start = datetime.fromtimestamp(df.index[0])
end = datetime.fromtimestamp(df.index[-1])
print(start, end)
```
